```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sy
```

```python
x = np.linspace(-1, 3, 1000)
def f(x): return x**2
```

```python
plt.plot(x, f(x))
plt.axhline(color='black')
plt.fill_between(x, f(x), where=[ (x>0) and (x<2) for x in x ])
```

```python
x = sy.Symbol('x')
```

```python
sy.integrate(f(x), (x, 0, 2))
```

```python
def g(x): return x**4
```

```python
x = np.linspace(-1, 1.5, 100)
plt.plot(x, f(x))
plt.plot(x, g(x))
plt.fill_between(x, f(x), g(x), where=[ (x>0) and (x<1) for x in x ])
```

```python
x = sy.Symbol('x')
sy.integrate(g(x) -f(x) , (x, 0, 1))
```
