# Superficie de revolucao 3D


```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

def f(x, y):
    return x**2 + y**2

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig = plt.figure()

ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Superfície de Revolução: $f(x, y) = x^2 + y^2$')
plt.show()
```

# Cone de revolucao 3d


```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Parâmetros do triângulo retângulo
base = 4.0
altura = 6.0

# Criação do cone de revolução
theta = np.linspace(0, 2 * np.pi, 100)
z = np.linspace(0, altura, 100)
Z, Theta = np.meshgrid(z, theta)  # Criação da matriz 2D para Z e Theta
X = base * np.cos(Theta)
Y = base * np.sin(Theta)

# Plotagem
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, alpha=0.7)

# Configurações adicionais
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Cone de Revolução')
plt.show()
```
