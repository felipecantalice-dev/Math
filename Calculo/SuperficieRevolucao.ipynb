```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Função
def f(x):
    return np.full_like(x,3)

# Intervalo
x = np.linspace(0, 5, 100)
```

```python
# Gráfico 2D
plt.figure(figsize=(7, 5))
plt.plot(x, f(x), label="f(x)=3")
plt.fill_between(x, f(x), color="skyblue", alpha=0.4)
plt.title("Gráfico 2D")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Gráfico da superfície de revolução
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
x = np.linspace(0, 5, 100)
theta = np.linspace(0, 2.*np.pi, 100)
x, theta = np.meshgrid(x, theta)
Y = f(x) * np.cos(theta)
Z = f(x) * np.sin(theta)
ax.plot_surface(x, Y, Z, rstride=5, cstride=5, color='k', edgecolors='w', linewidth=0.5)
ax.set_title("Superfície de Revolução")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
plt.show()
```
