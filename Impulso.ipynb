```python
# Impulso
Uma força mantida durante um longo período resulta maior momentum Quanto maior o impulso, maior o variação de momentum
```

## Forca constante

Aplicando a formula
$I=F*\Delta t$


```python
import numpy as np
import pandas as pd
import seaborn as sns

massa = 10 # kg
velocidade_inicial = 5 # m/s
velocidade_final = 10 # m/s

impulso = massa * (velocidade_final - velocidade_inicial)

print(f"Impulso: {impulso} kg m/s")

df = pd.DataFrame({
    "Massa": [massa],
    "Velocidade inicial": [velocidade_inicial],
    "Velocidade final": [velocidade_final],
    "Impulso": [impulso]
})

sns.barplot(x="Massa", y="Impulso", data=df)

sns.xlabel("Massa (kg)")
sns.ylabel("Impulso (kg m/s)")
```

```python
import matplotlib.pyplot as plt

massa = 10.0  # kg
forca = 5.0  # N
tempo_inicial = 0.0  # s
tempo_final = 10.0  # s

# Cálculo do impulso
impulso = forca * (tempo_final - tempo_inicial)

print(f"Impulso dentro do intervalo [0, 10]: {impulso} kg m/s")

# Criando dataframe
df = pd.DataFrame({
    "Tempo": np.linspace(tempo_inicial, tempo_final, 100),
    "Força": np.full(100, forca),
    "Impulso": np.full(100, impulso),
})

# Plotando gráfico da força
plt.plot(df["Tempo"], df["Força"], label="Força")
plt.xlabel("Tempo (s)")
plt.ylabel("Força (N)")

# Plotando gráfico do impulso
plt.plot(df["Tempo"], df["Impulso"], label="Impulso")
plt.xlabel("Tempo (s)")
plt.ylabel("Impulso (N s)")

plt.legend()
plt.show()
```

## Forca variavel
* Calcular a area do grafico
$I = Area$

Area do trapezio $\frac{(B+b)}{2}*h$


```python
import numpy as np
import pandas as pd
import seaborn as sns

massa = 10  # kg
tempo_inicial = 0  # s
tempo_final = 10  # s

# Função para definir a força variável
def forca(t):
    return 10 * t

# Cálculo do impulso usando integração numérica
tempos = np.linspace(tempo_inicial, tempo_final, 100)  # Define times explicitly
impulso = np.trapz(forca(tempos), tempos) * massa  # Pass 'tempos' to forca()

print(f"Impulso: {impulso} kg m/s")

# Criando dataframe
df = pd.DataFrame({
    "Tempo": tempos,  # Use the defined 'tempos'
    "Força": [forca(t) for t in tempos],  # Utilize 'tempos' here as well
})

# Plotando gráfico da força
sns.lineplot(x="Tempo", y="Força", data=df)

# Adicionando título e rótulos aos eixos
sns.set_title("Força variável")
sns.xlabel("Tempo (s)")
sns.ylabel("Força (N)")

# Plotando gráfico do impulso
sns.barplot(x="Tempo", y="Impulso", data=df)

# Adicionando título e rótulos aos eixos
sns.set_title("Impulso")
sns.xlabel("Tempo (s)")
sns.ylabel("Impulso (kg m/s)")
```
