```python
import matplotlib.pyplot as plt

x_data = [1, 2, 3, 4, 5]
y_data = [2, 4, 5, 8, 10]

plt.plot(x_data, y_data)
plt.show()
```

```python
def candidate_function(x, m, b):
  return m * x + b
```

```python
from scipy.optimize import curve_fit

popt, pcov = curve_fit(candidate_function, x_data, y_data)
```

```python
y_predicted = candidate_function(x_data, popt[0], popt[1])

plt.plot(x_data, y_data, label="Actual data")
plt.plot(x_data, y_predicted, color="red", label="Fitted function")
plt.legend()
plt.show()
```
